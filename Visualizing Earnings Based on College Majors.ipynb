# Analyzing pay for college majors

We will be analyzing a dataset on the median pay for specific college majors for students who graduated between 2010 and 2012. We also have different columns to aggregrate the data with. Below are the listed columns that will assist with analysis:

- Rank - Rank by median earnings (the dataset is ordered by this column).
- Major_code - Major code.
- Major - Major description.
- Major_category - Category of major.
- Total - Total number of people with major.
- Sample_size - Sample size (unweighted) of full-time.
- Men - Male graduates.
- Women - Female graduates.
- ShareWomen - Women as share of total.
- Employed - Number employed.
- Median - Median salary of full-time, year-round workers.
- Low_wage_jobs - Number in low-wage service jobs.
- Full_time - Number employed 35 hours or more.
- Part_time - Number employed less than 35 hours.

The data was released by the [American Community Survey](https://www.census.gov/programs-surveys/acs/). The news website [FiveThiryEight](https://fivethirtyeight.com/) cleaned the data and can be downloaded [here](https://github.com/fivethirtyeight/data/tree/master/college-majors)

# Initial Exploration

```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix

%matplotlib inline

recent_grads = pd.read_csv('recent-grads.csv')

# Initial Exploration first row, head, tail
print(recent_grads.iloc[0])
print('\n' * 2)
print(recent_grads.head())
print('\n' * 2)
print(recent_grads.tail())
```

```python
recent_grads.describe()
```

```python
recent_grads = recent_grads.dropna()
recent_grads.describe()
```

Removed null rows to prevent `metplotlib` from erroring out. It seems that only one row was null and thus removed.

My initial assessment by the current scheme of the data is that the wage gap (and major ranks) are very much split between gender. Where the top 5 majors all have less than 1/3 portion of the major share as women and the bottom 5 majors have over 2/3 share of women in the major.
